# Loading and Preprocessing

```python
from sklearn.datasets import load_breast_cancer

# Load the dataset
breastcancer_data = load_breast_cancer()
breastcancer_data
```

```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
df = pd.DataFrame(breastcancer_data.data, columns= breastcancer_data.feature_names)
df
```

```python
# adding target column

df["diagnosis"]= breastcancer_data.target
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.groupby("diagnosis").mean()
```

```python
x = df.drop(columns = "diagnosis", axis=1)
y = df["diagnosis"]
```

```python
x
```

```python
y
```

splitting the data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y)
```

# Logistic Regression Model

```python
logreg = LogisticRegression()
```

```python
logreg.fit(x_train, y_train)
```

```python
y_pred = logreg.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
```

```python
print("Accuracy score = ",accuracy)
```

```python
cr = classification_report(y_test, y_pred)
print("Classification Report:")
print(cr)
```

# Random forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y)
```

```python
clf = RandomForestClassifier()
```

```python
# Train the model

clf.fit(x_train, y_train)

#prediction

y_pred = clf.predict(x_test)
y_pred
```

```python
y_test
```

```python
#Evaluation

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy score = ",accuracy)
```

```python
cr = classification_report(y_test, y_pred)
print("Classification Report:")
print(cr)
```

```python
con = confusion_matrix(y_test, y_pred)
import matplotlib.pyplot as plt
import seaborn as sns
sns.heatmap(con, annot = True)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

# Model Comparison

After trying out different machine learning models, it is found that Random forest classifier has the highest f1-score. It combines the precision and recall scores of the model. The accuracy metric computes how many times a model made a correct prediction across the entire dataset.
Here we have Random forest classifier with 0.97 & 0.98 f1-score while the second best figures are for Decision tree classifier with 0.93 & 0.95.
